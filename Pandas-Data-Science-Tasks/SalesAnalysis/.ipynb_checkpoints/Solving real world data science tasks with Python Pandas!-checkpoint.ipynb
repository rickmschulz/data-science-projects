# Sales Analysis

#### Import Necessary Libraries

```python
import pandas as pd
import glob
```

#### Merging 12 months of Sale Data into a single Dataframe

```python
sales_months = glob.glob('./Sales_Data/*.csv')

df = pd.concat([pd.read_csv(df) for df in sales_months], ignore_index=True)

df.to_csv('all_months_data.csv', index=False)
```

```python
df
```

## Cleaning Up the Data

#### Drop Rows of NaN

```python
# Visualizing NaN cells
# nan_df = df[df.isna().any(axis=1)]
# nan_df

# Removing NaN rows
df.dropna(how='all', inplace=True)
```

#### Find 'Or' and Remove Rows

```python
df = df.loc[df['Order Date'].str[:2] != 'Or'].copy()
df
```

#### Converting Columns to Correct Data Types

```python
df = df.astype({'Quantity Ordered': 'int', 'Price Each': 'float'})
```

## Augment Data with Aditional Columns

#### Add Month Column

```python
# mask = all_months_data['Order Date'].str[:2].astype('int32')
# all_months_data['Month'] = None
# all_months_data.loc[mask, 'Month']
# all_months_data

df['Month'] = df['Order Date'].str[:2]
df = df.astype({'Month': 'int'})
df.dtypes
```

#### Add Total Column

```python
df.insert(4, 'Total', None)
df['Total'] = df['Quantity Ordered'] * df['Price Each']
df
```

#### Add City Column

```python
df.insert(len(df.columns)-2, 'City', None)
```

```python
# 1 Method
# cities = df['Purchase Address'].str.extract(r'(?<=,\s)([A-Za-z]+.*)(?=,)')
# df['City'] = cities

# 2 Method
# df['City'] = df['Purchase Address'].str.extract(r'(?<=,\s)([A-Za-z]+.*)(?=,)')

# 3 Method
# df.loc[:, 'City'] = cities.iloc[:, 0]
# df

# 4 Method
def get_city(address):
    return address.split(', ')[1]
    
    
def get_state(address):
    return address.split(', ')[2].split(' ')[0]


df['City'] = df['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
```

###  Question 1: What was the best month for sales? How much was earned that month?

```python
# month_sale = []
# for i in range(1, 13):
#     x = df.loc[df['Month'] == i].copy()
#     total = x['Total'].sum()
#     month_sale.append((total, i))
# print(month_sale)
# print(f'Best month for sale was {max(month_sale)[1]} and was made ${max(month_sale)[0]}.')

results = df.groupby('Month').sum()['Total']
results
```

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

# Resizing Graph
# plt.figure(dpi=150)
plt.figure(figsize=(10, 6))

months = list(range(1, 13))
months = [datetime.strptime(str(x), "%m").strftime("%b") for x in months]

plt.bar(months, results, label='Net Worth') 
plt.legend()

plt.xticks(months)
# plt.yticks(np.arange(max(results), step=max(results)/12))
plt.ylabel('Total Sales in USD ($)')
plt.xlabel('Month')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

### Question 2: What city had the highest number of sales?

```python
results1 = df.groupby(by=['City'])['Total'].sum().sort_values(ascending=False)
results1
```

```python
# cities = df['City'].unique()
# cities.sort()
# list(cities)
# cities = [x[:5]+'.' for x in cities if len(x) > 4]

cities = list(results1.index)
# cities = [x[:5]+'.' for x in cities if len(x) > 4]
cities
```

```python
# plt.figure(dpi=200)
plt.figure(figsize=(10, 6))

bars = plt.bar(cities, results1)

patterns = ['/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*']
for bar in bars:
    if len(patterns) == 0:
        break
    bar.set_hatch(patterns.pop(0))

plt.ylabel('Total Sales in USD ($)')
plt.xlabel('City')
plt.xticks(list(range(len(cities))), rotation='vertical')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

### Question 3: What time should we display advertisements to maximize likelihood of customer's buying product?

### Question 4: What products are most often sold together?

```python
# x = df.groupby(by=['Order ID', 'Product', 'Quantity Ordered']).count().sort_values('Order ID', ascending=True)
# x.head(50)

dfa = df[df['Order ID'].duplicated(keep=False)].copy()
dfa.head()
```

```python
dfa.groupby('Order ID')['Product'].transform(lambda x: ', '.join(x))
```

```python
dfa['Grouped Products'] = dfa.groupby('Order ID')['Product'].transform(lambda x: ', '.join(x))
dfa.head()
```

```python
dfa = dfa.loc[:, ['Order ID', 'Grouped Products']].drop_duplicates()
dfa
```

#### Method 1

```python
# 1 Method
dfb = dfa['Grouped Products'].value_counts().reset_index()
dfb.head(10)
```

```python
dfb.rename(columns={'Grouped Products': 'Count', 'index': 'Grouped Products'})
```

```python
dfb.columns
```

#### Method 2

```python
# 2 Method ---> Method did not gave the same results as Method 1!
from itertools import combinations
from collections import Counter

count = Counter()

for row in dfa['Grouped Products']:
    row_list = row.split(', ')
    count.update(Counter(combinations(row_list, 2)))
    
count.most_common(10)
```

#### Plotting

```python
plt.figure(figsize=(14, 7))

bars = plt.bar(dfb.loc[:9, 'index'] , dfb.loc[:9, 'Grouped Products'])

patterns = ['/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*']
legends = list(dfb.loc[:9, 'index'])
colors = ['red', 'yellow', 'orange', '#cd9575', '#5d8aa8', '#00ffff', '#fae7b5', '#fb607f', 'green', '#8a2be2']

for bar in bars:
    bar.set_label(legends.pop(0))
    bar.set_color(colors.pop(0))

for bar in bars:
    if len(patterns) == 0:
        break
    bar.set_hatch(patterns.pop(0))

plt.legend(fontsize='large')    

bar.axes.get_xaxis().set_visible(False)
plt.ylabel('Counts', labelpad=20, fontsize='large')
plt.xlabel('Grouped Products')
plt.xticks(list(range(10)))
plt.yticks(fontsize='large')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

### Question 5: What product sold the most? Why do you thing it sold the most?

```python
dfc = df.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False).reset_index()
dfc
```

```python
# plt.figure(figsize=(12, 7))

products = dfc.loc[:, 'Product']
quantity_ordered = dfc.loc[:, 'Quantity Ordered']

bars = plt.bar(products, quantity_ordered)

patterns = ['/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*']

for bar in bars:
    if len(patterns) == 0:
        break
    bar.set_hatch(patterns.pop(0))

plt.ylabel('Quantity Ordered', labelpad=20, fontsize='large')
plt.xlabel('Products', fontsize='large')
plt.xticks(rotation='vertical')
plt.yticks(fontsize='large')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

#### Correlation with Product Prices 

```python
dfc.values[:, 0]
# dfc.values
```

```python
prices = df.groupby('Product')['Price Each'].mean()
prices
```

```python
prices_sorted = []
for product in dfc.values[:, 0]:
    print(product, prices[product])
    prices_sorted.append(prices[product])

prices_sorted
```

```python
fig, ax1 = plt.subplots(figsize=(14, 7))

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices_sorted, 'b--', linewidth=3)


ax1.set_ylabel('Quantity Ordered', labelpad=20, fontsize='large')
ax1.set_xlabel('Products', fontsize='large')
ax1.set_xticks(list(range(len(products))), products, rotation='vertical')

ax2.set_ylabel('Product Price (US$)', fontsize='large')

plt.show()
```
